```python
star_type= input("1) what is the type of the star? (e.g. G-type, K-type, M-type, F-type, A-type): ")
star_age= float(input("2) what is the age of the star? (in billions of years): "))
distance= float(input("3) what is the distance between the planet and the mother star? (in Astronomical Units e.g. 2 AU) "))
radius = float(input("4) What is the planet's radius (in Earth radii)? "))
temperature = float(input("5) What is the surface temperature (¬∞C)? "))
atm_density = float(input("6) Atmospheric pressure (in bars, Earth = 1 bar): "))
mag_field = input("7) Does the planet have a magnetic field? (yes/no): ")

print("\nThank you! Your answers have been recorded.")
print("Star type:", star_type)
print("Star age:", star_age, "billion years")
print("Planet distance:", distance, "AU")
print("planet radius:", radius)
print("temperature:", temperature)
print("atmosphere density=", atm_density)
print("magnetic field:", mag_field)

```

```python
habitable = True

#star type---
if star_type.lower().startswith("m"):
    print("‚ö†Ô∏è M-type star: may have strong radiation and flares.")
elif star_type.lower().startswith(("o", "b")):
    print("‚ùå O/B type star: too short-lived for life.")
    habitable = False

#star Age-----

if star_age < 0.5:
  print("star too young, system unstable")
  habitable = False

# ---- Dynamic Habitable Zone based on star type ----
star_type_luminosity = {
    "O": 100000, "B": 10000, "A": 25,
    "F": 6, "G": 1, "K": 0.4, "M": 0.05
}

# Get luminosity; default to 1 if unknown
lum = star_type_luminosity.get(star_type.upper(), 1)

# Calculate inner and outer habitable zone (AU)
inner_hz = 0.95 * (lum ** 0.5)
outer_hz = 1.37 * (lum ** 0.5)

print(f"Habitable zone for {star_type}-type star: {inner_hz:.2f} AU to {outer_hz:.2f} AU")

# Check if planet distance is within HZ
if distance < inner_hz or distance > outer_hz:
    print("‚ùå Planet is outside the habitable zone.")
    habitable = False
else:
    print("‚úÖ Planet is within the habitable zone.")

#Radius---------
if radius < 0.5:
    print(" Too small; cannot hold atmosphere.")
    habitable = False
elif radius > 1.8:
    print("Too big; likely a mini-Neptune or super thick atmosphere.")
    habitable = False
elif radius == 1:
  print("Congrats! you found an earth rocky twin!")

  #Temperature -------

if temperature<-20 or temperature>60:
  print("Biological temperature range failed [-20, 60 C] and not stable for liquid water")
  habitable = False
elif temperature > 50:
    print("‚ö†Ô∏è Very hot; life possible but only for thermophiles.")

# ---- Atmosphere Pressure ----
if atm_density < 0.5:
    print("‚ùå Atmosphere too thin. Water will evaporate.")
    habitable = False
elif atm_density > 5:
    print("‚ùå Atmosphere too thick. Extreme greenhouse effect.")
    habitable = False

# ---- Magnetic Field ----
if mag_field.lower() == "no":
    print("‚ö†Ô∏è No magnetic field: atmosphere vulnerable to stellar wind.")

    # ---- Final Result ----
print("\n-------------------------------------------")
if habitable:
    print("üåç RESULT: The planet is *potentially habitable*!")
else:
    print("‚ùå RESULT: The planet is NOT habitable with current conditions.")
print("-------------------------------------------")


```
